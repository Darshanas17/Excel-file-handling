```python
import openpyxl 
import os.path 
from openpyxl import load_workbook 
import time 
import random 

path = input('Enter an EXCEL FILE path: ')  

if os.path.exists(path):  
    print('\n')  
    print('Opening file...')  
    print('\n')
    
    t = random.randint(3, 5)  
    time.sleep(t) 

    wb_obj = openpyxl.load_workbook(path) 
    sheet_obj = wb_obj.active 

    row = sheet_obj.max_row 
    column = sheet_obj.max_column 

    print("\n--------List of students--------") 
    for i in range(1, row + 1): 
        cell_obj = sheet_obj.cell(row=i, column=1) 
        cell_obj2 = sheet_obj.cell(row=i, column=2) 
        print(cell_obj.value, '    ', cell_obj2.value) 

    print('\n') 
    a = int(input('Enter an SL.no to print student information:  ')) 
    print('\n') 
    for i in range(1, column + 1): 
        cell_obj = sheet_obj.cell(row=1, column=i) 
        print(cell_obj.value, end="             ") 
    print('\n') 

    for i in range(1, column + 1): 
        cell_obj = sheet_obj.cell(row=a + 1, column=i) 
        print(cell_obj.value, end="              ") 

    print('------------------------------------------------------------------------------------------------------------------------------') 

    while True:  
        print('\n') 
        print('Do you want to update any data?\nPress 1 to Update\nPress 2 to Exit') 
        ans = input() 
        print('---------------------------------------------------------------------------------------------------------------') 

        if ans == '1': 
            workbook = load_workbook(path) 
            sheet = workbook.active 
            ch = int(input('What do you want to modify?\nPress 1 to modify NAME\nPress 2 to modify '
                            'USN\nPress 3 to modify BRANCH\nPress 4 to modify SEMESTER\nPress 5 to modify PHONE '
                            'NUMBER\nPress 6 to modify BLOOD GROUP\n')) 

            ch = ch + 1 
            if ch == 1: 
                x = 'A' 
            elif ch == 2: 
                x = 'B' 
            elif ch == 3: 
                x = 'C' 
            elif ch == 4: 
                x = 'D' 
            elif ch == 5: 
                x = 'E' 
            elif ch == 6: 
                x = 'F' 
            elif ch == 7: 
                x = 'G' 

            if ch == 2: 
                str2 = input('Enter a NAME: ') 
                str3 = 'NAME' 
            elif ch == 3: 
                str2 = input('Enter a USN: ') 
                str3 = 'USN' 
            elif ch == 4: 
                str2 = input('Enter a BRANCH: ') 
                str3 = 'BRANCH' 
            elif ch == 5: 
                str2 = input('Enter a SEMESTER: ') 
                str3 = 'SEMESTER' 
            elif ch == 6: 
                str2 = input('Enter a PHONE NUMBER: ') 
                str3 = 'PHONE NUMBER' 
            elif ch == 7: 
                str2 = input('Enter a BLOOD GROUP: ') 
                str3 = 'BLOOD GROUP' 
            else: 
                print('                                           Invalid Choice') 
                print('---------------------------------------------------------------------------------------------------------------') 
                break 

            b = a + 1 
            y = str(b) 
            str1 = x + y 

            sheet[str1] = str2 

            workbook.save(path) 

            print('---------------------------------------------------------------------------------------------------------------') 
            print(str3, 'is Modified to ->', str2) 
            print('Successfully Updated!') 

            print('\n') 
            print('Updated data:') 

            wb_obj = openpyxl.load_workbook(path) 
            sheet_obj = wb_obj.active 

            for i in range(1, column + 1): 
                cell_obj = sheet_obj.cell(row=1, column=i) 
                print(cell_obj.value, end="             ") 
            print('\n') 

            for i in range(1, column + 1): 
                cell_obj = sheet_obj.cell(row=a + 1, column=i) 
                print(cell_obj.value, end="              ") 
        elif ans == '2': 
            print('**************************************************************************') 
            print('**************************************************************************') 
            print('**************************************************************************') 
            break 
        else: 
            print('                                           Invalid Choice') 
            print('---------------------------------------------------------------------------------------------------------------') 
            break 
else: 
    print('Path not found!')
```
